# Conditional Statement

|Part|Purpose|Condition Evaluated|Action if True|
|---|---|---|---|
|`if`| First condition check | First condition to be evaluated | Executed block if condition is true |
|`elif`| Subsequent condition check | Next condition, `if` is false | Executes block if this condition is true |
| `else` | Fallback if no condition are met | None | Executes block if all above conditions are false |

# OBJECTS: THE DATA RECORDS
**Object:** A data record with fields; an instance of a class. Everything in python can be defined as object. The variable, value of a variable, function, data, etc can be called and defined as object.

## Variables: The Reference to Objects
**Variable:** A defined name that references an objects. Variable names itself point to an object. id function gives us identification number for an object to see if it's correalted to another function.

````python
job_title = "Data Analyst"
id(job_title)
````

## Functions: The Manipulators of Objects
**Function:** A reusable piece of code that performs a specific task. In order to run a function we need to add parenthesis and pass something as argument. It can be built in or build our own function.

Below code block would have print() which is a built in function and we will define a new function. def is used to define a new function followed by the function name.

````python
print("I am trying something new!")
def test_one():
  return "I am trying something new!"
test_one()
````

````python
job_title = "Data Analyst"
job_location = "India"
job_salary = 1500000
def display_info(title, location, salary):
  return print(f"Job: {title}\nLocation: {location}\nSalary: ₹{salary:,.0f}")
display_info(job_title, job_location, job_salary)
````

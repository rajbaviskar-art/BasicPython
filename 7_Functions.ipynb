##User Defined Functions


Below we calculate the total salary of a person.

total_salary = base_salary * (1 + bonus_rate)

```python
base_salary = 100000
bonus_rate = 0.1

def calculate_salary(base_salary, bonus_rate):
  total_salary = base_salary * (1 + bonus_rate)
  return total_salary
```

```python
calculate_salary(150000, 0.04)
```

When we know that the rate is not changing.

```python
def calculate_salary(base_salary, bonus_rate = 0.1):
  total_salary = base_salary * (1 + bonus_rate)
  return total_salary
```

```python
calculate_salary(20000)
```

It can also override the predefined rates if we add rate while calling that function.

```python
calculate_salary(1000000,0.5)
```

##Lambda Functions

```python
mul_two = lambda x: x * 2
```

```python
mul_two(3)
```

```python
(lambda x, y: x * 2 + y)(3, 2)
```

```python
data_salaries = [4_50_000, 9_00_000, 12_00_000, 20_00_000 ]

def calculate_salary(base_salary, bonus_rate = 0.1):

  return base_salary * (1 + bonus_rate)

total_salary_list = [calculate_salary(salary) for salary in data_salaries]

total_salary_list
```

Replacing the above function in lambda functions

```python
data_salaries = [4_50_000, 9_00_000, 12_00_000, 20_00_000 ]

total_salary_list = [(lambda x: x * 1.1)(salary) for salary in data_salaries]

total_salary_list
```

Find jobs that have Python and are Remote

```python

jobs_data = [
    {'job_title': 'Data Scientist',  'job_skills': ['Python', 'Machine Learning'], 'remote': True},
    {'job_title': 'Data Analyst',  'job_skills': ['Excel', 'SQL'], 'remote': False},
    {'job_title': 'Machine Learning Engineer', 'job_skills': ['Python', 'TensorFlow', 'Keras'], 'remote': True},
    {'job_title': 'Software Developer', 'job_skills': ['Java', 'C++'], 'remote': True},
    {'job_title': 'Data Scientist', 'job_skills': ['R', 'Statistics'], 'remote': False}
]

list(filter(lambda job: job['remote'] and "Python" in job['job_skills'], jobs_data))
```

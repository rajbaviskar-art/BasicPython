# Data Clean up

The dataset we have been refering to has around 787686 records out of which there are only 33073 records for salary_rate which is not null. So we want to fill in the NaN values with some values like median. We also want to remove any duplicates values.

```python
#importing libraries
import pandas as pd
from datasets import load_dataset
from matplotlib import pyplot as plt

# Loading our dataset
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
# Checking for NaN values
df.info()
```

```python
# To fill in median data we first create a variable for it
median_salary_year = df['salary_year_avg'].median()
median_salary_hour = df['salary_hour_avg'].median()

# To fill NaN columns we will use fillna method
df_filled = df
df_filled['salary_year_avg'] = df['salary_year_avg'].fillna(median_salary_year)
df_filled['salary_hour_avg'] = df['salary_hour_avg'].fillna(median_salary_hour)
```

```python
df_filled.loc[:10,'salary_year_avg':'salary_hour_avg']
```

```python
# now we will drop some duplicates using drop_duplicates()

df_unique = df_filled
df_unique = df_unique.drop_duplicates()
```

```python
# Testing if the duplicates were dropped

print('Length of orginal df: ', len(df_filled))
print('Length of df after dropping duplicates: ', len(df_unique))
print('Rows Dropped: ', len(df_filled) - len(df_unique))
```

```python
# We actually want to specify what specific columns to look after while dropping duplicates

df_unique = df_unique.drop_duplicates(subset = ['job_title', 'company_name'])

# Testing if the duplicates were dropped

print('Length of orginal df: ', len(df_filled))
print('Length of df after dropping duplicates: ', len(df_unique))
print('Rows Dropped: ', len(df_filled) - len(df_unique))
```

# Pivot Tables

We will check how median salaries look for top 6 countries look like for the top 3 job titles

```python
#importing libraries
import pandas as pd
from datasets import load_dataset
from matplotlib import pyplot as plt

# Loading our dataset
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
# Top countries

top_countries = df['job_country'].value_counts().head(6).index

#Creating the pivot
df_job_country_salary = df.pivot_table(
    values = 'salary_year_avg',
    index = 'job_country',
    columns = 'job_title_short',
    aggfunc = 'median'
)

df_job_country_salary = df_job_country_salary.loc[top_countries]

job_titles = ['Data Analyst', 'Data Engineer', 'Data Scientist']

df_job_country_salary = df_job_country_salary[job_titles]

df_job_country_salary

df_job_country_salary.plot(kind = 'bar')
plt.xlabel('Country')
plt.ylabel('Median Salary ($)')
plt.title('Median Salary by Country and Job Title')
plt.legend(title = 'Job Title')
plt.xticks(rotation = 45, ha = 'right')
plt.show()
```

# Index Management


```python
#importing libraries
import pandas as pd
from datasets import load_dataset
from matplotlib import pyplot as plt

# Loading our dataset
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

Three Methods to cover
- df.reset_index()
- df.set_index()
- df.sort_index()

**Reset Index**

```python
df_usa = df[df['job_country'] == 'United States']
```

```python
df_usa
```

```python
# Above we see the job_index is not in order so we can reset to the df
df_usa.reset_index(inplace = True)
df_usa
```

**Set Index**

```python
# We can set a particular column as our index
df_usa.set_index('job_index', inplace = True)
df_usa
```

**Sort Index**

```python
# We can return a index in sorted usually used when we don't have a index title
median_pivot = df_usa.pivot_table(
    values = 'salary_year_avg',
    index = 'job_title_short',
    columns = 'job_country',
    aggfunc = ['median', 'min', 'max']
)

median_pivot.sort_index(inplace = True)

median_pivot
```

# Exercise

What is the demand of certain job titles over a period of a year



```python
#importing libraries
import pandas as pd
from datasets import load_dataset
from matplotlib import pyplot as plt

# Loading our dataset
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
# Create a new dataframe for India specific jobs
df_india = df[df['job_country'] == 'India'].copy()
```

```python
#Adding a new column with month of the posting date
df_india['job_posted_month'] = df_india['job_posted_date'].dt.strftime('%B')

df_india
```

```python
# Lets create a pivot
df_india_pivot = df_india.pivot_table(
    index = 'job_posted_month',
    columns = 'job_title_short',
    aggfunc = 'size'
)
```

```python
# Working with indexes to sort the output in correct month order
df_india_pivot.reset_index(inplace = True)
df_india_pivot['month_no'] = pd.to_datetime(df_india_pivot['job_posted_month'], format = '%B').dt.month
df_india_pivot.sort_values('month_no', inplace = True)
df_india_pivot.set_index('job_posted_month', inplace = True)
df_india_pivot.drop(columns = ['month_no'], inplace = True)
df_india_pivot
```

```python
# Getting the top 3 job roles
top_3 = df_india['job_title_short'].value_counts().head(3)
top_3 = top_3.index.tolist()
```

```python
df_india_pivot[top_3].plot(kind = 'line')
plt.xlabel('Month (2023)')
plt.ylabel('Job Count')
plt.title('Monthly Job Posting for top data jobs in India')
plt.legend(title = 'Job Title')
plt.show()
```

# Data Containers

There are 4 types of data containers:
1. Lists
2. Dictionaries
3. Sets
4. Tuples

|Feature|List|Dictionary|Set|Tuple|
|-------|----|----------|---|-----|
|Usage| For a collection of ordered items | For key-value pairs | For unique items | For fixed data|
|Syntax|`[item1,item2,...]`|`{'key1':value1, 'key2':value2,...}`|`{item1,item2,...}`|`(item1,item2,...)` or `item1,item2,...`|
|Type of data | Sequence | Mapping | Set | Sequence|
|Order |Ordered | Unordered | Unordered | Ordered |
|Indexing | Yes (by index) | Yes (by key) | No | Yes (by index) |
| Duplicate Values | Allowed | Vales can be duplicated, keys cannot | Not allowed | Allowed |
| Mutability | Mutable | Mutable | Mutable | Immutable |

##Lists

List is donated by square brackets []. Can contain any data type in it.

```python
job_skills = ['SQL', 'Tableau' , 'Excel']
job_skills
```

**Append:** Adds new object at the end of the list

```python
job_skills.append('Python')
job_skills
```

**Remove:** Removes the first occurance of the object from the list.

```python
job_skills.remove('Tableau')
job_skills
```

**Length:** Returns the number of items in a container.

```python
len('Python')
```

```python
len(job_skills)
```

###Indexes of list
Each object in list is indexed using numbers. List indexing starts from 0. For job_skills list defined above the "SQL" is indexed at 0, "Excel" at 1 and "Python" at 2.

```python
job_skills[1]
```

**insert:** Insert object before index.

```python
job_skills = ['SQL', 'Excel', 'Python']
job_skills.insert(2, 'Tableau')
job_skills
```

**pop:** Remove and return the item at index which is defaulted at last.

```python
job_skills.pop(2)
```

```python
job_skills
```

```python
job_skills.pop()
```

```python
job_skills
```

**Slicing:** Slicing is used to access multiple indexes in a list

Syntax: `list[start:end:step]`

`start`: The starting index (inclusice).

`end`: The ending index (exclusice).

`step`: Steps to take between items.

```python
job_skills = ['SQL','Excel','Python']
job_skills[0:2]
```

```python
job_skills[:]
```

```python
job_skills[:1]
```

```python
job_skills[::2]
```

```python
raj_skills = ['Python', 'BigQuery', 'R']
mansi_skills = ['Python', 'SQL', 'Looker']
all_skills = raj_skills + mansi_skills
all_skills
```

```python
all_skills[::2]
```

###Unpacking

Assign each value in an iterable to a variable in a single statement.


```python
raj_skills = ['Python', 'BigQuery', 'R']
skill1, skill2, skill3 = raj_skills
print(skill1)
print(skill2)
print(skill3)
```

**Unpack or Star Operator:** Unpacks a list and assigns lists to a new variable.

```python
skill_concerned, *skill_not_concerned = raj_skills
print(skill_concerned)
print(skill_not_concerned)
```

##Sets

Very similar to list but instead of [] brackets we use {} brackets without colan to not make it a dictionary.

Sets are unique. If we convert a list to set it will remove duplicates if any. They are mostly designed for unique items.

Sets are not ordered.

Primary use case of set is to remove duplicate values.

```python
job_skills = {'tableau', 'SQL', 'Python', 'Statistics'}
job_skills
```

```python
job_skills.add('looker')
job_skills
```

```python
job_skills.pop()
```

```python
job_skills.pop('Statistics')
job_skills
```

```python
job_skills = {'tableau', 'SQL', 'Python', 'Statistics'}
job_skills.remove('Statistics')
job_skills
```

```python
skill_list= ['python', 'sql', 'statistics', 'tableau', 'python', 'sql', 'statistics', 'tableau']
list(set(skill_list))
```

##Tuples

This is similar to list as this is sequential data container. Data is added to it using (). Any iteams can be added to this. Tuples are Immutable we can't change the data in this. This is more applicable in software engineering where we don't want it to change.

Tuples are indexable.



```python
raj_skills = ('python', 'bigquery', 'r')
raj_skills
```

```python
raj_skills[0]
```

```python
raj_skills[0:2]
```

```python
raj_skills.append('sql')
```

We can still add new objects in sets by using + operator and adding objects from a different tuple. We are actually creating a entirely new object using this method and not updating the old object.

```python
raj_skills = ('python', 'bigquery', 'r')
mansi_skills = ('python', 'sql', 'looker')
all_skills = raj_skills + mansi_skills
all_skills
```

```python
raj_skils = ('python', 'bigquery', 'r')
new_skills = ('excel', 'power bi')
raj_skils += new_skills
raj_skils
```

```python
raj_skills = ('python', 'bigquery', 'r')
print(id(raj_skills))
new_skills = ('excel', 'power bi')
raj_skills += new_skills
raj_skills
id(raj_skills)
```

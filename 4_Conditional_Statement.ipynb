# Conditional Statement

|Part|Purpose|Condition Evaluated|Action if True|
|---|---|---|---|
|`if`| First condition check | First condition to be evaluated | Executed block if condition is true |
|`elif`| Subsequent condition check | Next condition, `if` is false | Executes block if this condition is true |
| `else` | Fallback if no condition are met | None | Executes block if all above conditions are false |

```python
if True:
  print("I am randomly testing if statement")
```

Below code will simply compare a job skill and applicant skill

```python
job_skill = 'Java'
applicant_skill = 'SQL'
years_experince = 6
if job_skill == applicant_skill:
  print("Skills Match")
```

Below code we added a elif statement to talk about years of experince.

```python
if job_skill == applicant_skill:
  print("Skills Match")
elif years_experince >= 5:
  print("Enough Experince; No SQL")
```

```python
job_skill = 'Java'
applicant_skill = 'SQL'
years_experince = 4
if job_skill == applicant_skill:
  print("Skills Match")
elif years_experince >= 5:
  print("Enough Experince; No SQL")
else:
  print("No Skill or Experince")
```

pass(...) is used to sometimes just fill in some data after if,elif statement to let the system run the function till we know what to do for that condition

```python
job_skill = 'Java'
applicant_skill = 'SQL'
years_experince = 4
if job_skill == applicant_skill:
  print("Skills Match")
elif years_experince >= 5:
  print("Enough Experince; No SQL")
else:
  pass
```

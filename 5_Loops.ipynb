##FOR Loops
Iterating over a sequence (list, range, string).Use a for loop when you have a definite set of data to process. For example, if you want to print every name in a list of users or square every number from 1 to 10 using the range() function.


```python
numbers = [0, 1, 2, 3, 4, 5]
for number in numbers:
  print(number)
```

```python
characters = 'PYTHON'
for character in characters:
  print(character)
```

The below code will be used to hire someone with more than 5 years of experince.

```python
years_experince  = {
    'RAJ': 4,
    'MANSI': 6,
    'ANIRUDHH': 5,
    'SATYAM': 2
}

for person in years_experince:
  print(person)
```

```python
for value in years_experince.values():
  print(value)
```

```python
for key,value in years_experince.items():
  print(key,value)
```

```python
for key,value in years_experince.items():
  if value > 5:
    print(key, 'has', value, 'years of experince')
```

```python
for _ in range(100):
  print(_)
```

##While Loops
Repeating code as long as a condition is true. Use a while loop when the stopping point depends on something unpredictable, such as waiting for a user to type "quit," monitoring a sensor until it reaches a certain temperature, or handling a dynamic event.

```python
count = 1
while count < 5:
  print(count)
  count += 1
```

```python
years_experince  = {
    'RAJ': 4,
    'MANSI': 6,
    'ANIRUDHH': 5,
    'SATYAM': 2,
    'SAWAI': 7
}

years = 0
index = 0
years_list = list(years_experince.items())
while years < 5:
  key, years = years_list[index]
  if years >= 5:
    print(key, 'has', years, 'years of experince')
  index += 1
```

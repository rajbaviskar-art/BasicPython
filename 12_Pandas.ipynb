#Pandas

- Pandas is a Python library used for working with data sets.
- It lets us analyze, clean, explore and manipulate data.
- Pandas primarily uses two data structures to store data:
- Series: A one-dimensional array with data labels, called its index, capable of holding any data type. It's like a column in a spreadsheet.
- DataFrame: A two-dimensional, mutable table with labeled axes (rows and columns). It resembles a spreadsheet or SQL table and can contain multiple Series objects of different data types.

##Pandas Exercise

```python
# Starting everything from scratch for better readibility
!pip install datasets
import pandas as pd
from datasets import load_dataset

dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
# Aggregated the job countries to check top 20 countries by count
df['job_country'].value_counts().head(20)
```

```python
#Checks if there are any records for a country
df['job_country'].isin(['India']).any()
```

```python
#Filtered for only India as country
india_jobs = df[df['job_country'] == 'India']
```

```python
india_jobs
```

```python
# There are NA values in salary column which we don't want
india_jobs = india_jobs[india_jobs['salary_year_avg'].notna()]
india_jobs
```

```python
india_jobs.groupby('job_title_short')['salary_year_avg'].agg(['median', 'min', 'max', 'count']).sort_values(by = 'median', ascending = False)
```

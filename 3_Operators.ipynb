## Order of Operations
1. **Parentheses:** Anything in parentheses is executed first.
2. **Exponents:** Operations involving exponents (including roots, indicated as fractional exponents).
3. **Multiplication and Division:** Processed from left to right.
4. **Addition and Subtraction:** Also processed from left to right.

```python
min_salary = 20_000 # _ is ignored by the interpreter
max_salary = 80_000

avg_salary = (min_salary + max_salary) / 2

print(avg_salary)
```

##Logical Operators
| Operator | Operator Type | Example |
|----------|---------------|---------|
| and | Logical AND | x and y |
| or | Logical OR | x or y |
| not | Logical NOT | not x |

```python
job_wfh = True
job_health_insurance = True
job_no_degree_mention = True

job_wfh and job_health_insurance and job_no_degree_mention
```

```python
job_wfh = True
job_health_insurance = False

job_wfh or job_health_insurance
```

```python
job_wfh = False
job_health_insurance = True
job_no_degree_mention = True

job_wfh or job_health_insurance or job_no_degree_mention
```

```python
job_wfh = False
job_health_insurance = False
job_no_degree_mention = True

job_wfh or job_health_insurance or not job_no_degree_mention
```

```python
if job_wfh or job_health_insurance or not job_no_degree_mention:
  print("Meets Condition")
else:
  print("Does not meet condition")
```

## Identity Operators

| Operator | Operator Type | Example |
|----------|---------------|---------|

If something is correlated or not. It checks if two things have same memory location.

```python
salary_raj = 10_00_000
salary_mansi = 10_00_000
salary_raj is salary_mansi
```

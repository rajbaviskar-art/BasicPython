Let's try importing a simple csv file within python and run some basic analysis

```python
file = open('/content/sample_data/california_housing_test.csv')

content = file.read()

file.close()
```

```python
import csv

data_dict = {}
for index, row in enumerate(csv.reader(content.strip().split('\n'))):
  if index == 0:
    # Initialize dictionary with column headers as keys
    for column in row:
      data_dict[column] = []
  else:
    # Append each element in the row to correct the list in dictionary
    for col_index, value in enumerate(row):
      data_dict[list(data_dict.keys())[col_index]].append(value)


data_dict
```

```python
sum(data_dict['total_rooms'])
```

```python
sum(float(room.replace(' ','')) for room in data_dict['total_rooms'] if room.replace(' ', '').replace('.',''). isdigit())
```

**Module:** A module is a single python file containing definitions and statements



**Package:** A collection of python modules under a common namespace. This is achieved by having a directory with a special `__init__.py` file.

**Library:** A collection of packages or modules that are grouped together to provide functionality that can be utilized by other programs.
